```python
def sumar(nro1 , nro2):
    print (nro1 + nro2)

def restar(nro1, nro2):
    print ("La resta es: ", nro1 - nro2)
    
def producto(nro1, nro2):
    print ("El producto es: ", nro1 * nro2)

def division(nro1 , nro2):
    print("La división es:", nro1 / nro2)    
```

```python
nro1 = int(input("Ing. el 1er. nro: "))
nro2 = int(input("Ing. el siguiente. nro: "))
operador = input("Ingresa un operador: + | - | * | /: ")
if operador == '+':
    sumar (nro1, nro2)
elif operador == '-':
    restar (nro1, nro2)
elif operador == '*':
    producto (nro1, nro2)
elif operador == '/':
    division (nro1, nro2)
else:
    print("No haz elegido bien !")
```

Leer las edades de N personas al azar, almacenarlas en una lista y determinar cuántas de esas personas son mayores de edad.

```python
def personasMayores(cantidad):
  import random
  lista = [] 
  cantPersonasMay = 0    # contar la cantidad de personas mayores de 18 años
  for m in range(1,cantidad):
    edad1 =  random.randint(1,100)  
    lista.append(edad1)  
    if edad1 >=18 :
      cantPersonasMay +=1  # cantPersonasMay = cantPersonasMay + 1
  print("La cant. de personas mayores a 18 años es: " , cantPersonasMay)
  print(lista)
```
