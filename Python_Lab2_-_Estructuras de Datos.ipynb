### 5. Funciones

```python
def esPrimo(numero):
  cantDiv=0
  for i in range (1,numero +1):
    if numero % i == 0:
      cantDiv +=1   # cantDiv = cantDiv + 1
  if cantDiv == 2:
    print("El numero ", numero, " es PRIMO")
  else:
    print("El numero ", numero, " NO es PRIMO")

def esPerfecto(numero):
  suma = 0
  for i in range(1,numero):
    if numero % i == 0:
      suma += i   # suma = suma + i
  if suma == numero:
     print("El numero ", numero, " es PERFECTO")
  else:
    print("El numero ", numero, " NO es PRIMO")

def factorial(numero):
    factorial= 1
    for i in range(numero,1,-1):
        factorial *= i # factorial = factorial * i
    print("El factorial es: " + str(factorial))   

opcion = 1
while opcion == 1:
    ejercicio = int(input("Escoge tu ejercicio -> 1:Primo, 2:Perfecto, 3: factorial"))
    if ejercicio > 3:
        print("No entiendes?!!! s칩lo pon el 1 o el 2 o el 3")
        continue
    numero = int(input("Ingresa el n칰mero: "))
    if ejercicio == 1:        
        esPrimo(numero)
    elif ejercicio == 2:    
        esPerfecto(numero)
    else:    
        factorial(numero)        
    opcion = int(input("Deseas continuar? 1:SI, 2: NO"))
    if opcion == 2:
        print("Hasta la vista baby")
        break
```
